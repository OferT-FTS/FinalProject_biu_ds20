```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

#  专转 转 砖 转 驻 (AAPL) 注 转 砖注
ticker = "AAPL"
df = yf.download(ticker, start="2020-01-01", end="2025-04-02")
df.to_csv(f"{ticker}.csv")
df = df[['Close']]  # 砖专 专拽 转 专 转
df.rename(columns={'Close': 'price'}, inplace=True)

#  转转  SARIMA
# 专转  注专 AR, MA, Seasonal AR - Seasonal MA
p_range = range(0, 5)
d_range = range(0, 5)
q_range = range(0, 5)
P_range = range(0, 5)
D_range = range(0, 5)
Q_range = range(0, 5)
s = 52  # 注转转 砖注转

# 驻拽爪 专爪转 SARIMA 专转 AIC
def run_sarima(df, p, d, q, P, D, Q, s):
    try:
        model = sm.tsa.statespace.SARIMAX(df['price'], order=(p, d, q), seasonal_order=(P, D, Q, s))
        results = model.fit(disp=False)
        return results.aic
    except:
        return np.inf  #  砖 砖, 专 住祝  转注  

# 驻砖 专 砖  转专 砖 驻专专 注 驻 AIC
best_aic = float('inf')
best_params = None

# 专砖 砖专转 转爪转
results_list = []

# 转 拽 转  砖 驻砖专
for p in p_range:
    for d in d_range:
        for q in q_range:
            for P in P_range:
                for D in D_range:
                    for Q in Q_range:
                        aic = run_sarima(df, p, d, q, P, D, Q, s)
                        print(f"SARIMA({p},{d},{q})x({P},{D},{Q},{s}) -> AIC: {aic}")

                        # 住驻转 转爪 专砖
                        results_list.append(((p, d, q, P, D, Q), aic))

# 爪专转 DataFrame 注 转爪转
results_df = pd.DataFrame(results_list, columns=['Parameters', 'AIC'])

# 爪转 转爪转
print("\nBest Parameters and AIC:")
print(results_df)

#  驻专专  转专
best_params = results_df.loc[results_df['AIC'].idxmin()]['Parameters']
print(f"\nBest Parameters: {best_params}")

#  专爪转 SARIMA 注 驻专专  转专
p, d, q, P, D, Q = best_params
model = sm.tsa.statespace.SARIMAX(df['price'], order=(p, d, q), seasonal_order=(P, D, Q, s))
results = model.fit(disp=False)

#   21 砖注转 拽
forecast_period = 21
forecast = results.predict(start=df.index[-1] + pd.Timedelta(weeks=1),
                           end=df.index[-1] + pd.Timedelta(weeks=forecast_period))

#  爪专转 爪专  注专 
forecast_dates = pd.date_range(start=df.index[-1], periods=forecast_period+1, freq='W')[1:]

#  驻住转 专祝 住专 +  + 驻转 专 专-住
plt.figure(figsize=(12,6))
plt.plot(df.index, df['price'], label="专 住专", color='blue')
plt.plot(forecast_dates, forecast, label="转转 21 砖注转 拽", color='red')

# 驻转 专 (95% Confidence Interval) - 砖 注专转 注转 住住拽
conf_int = results.get_forecast(steps=forecast_period).conf_int(alpha=0.05)
plt.fill_between(forecast_dates,
                 conf_int.iloc[:, 0], conf_int.iloc[:, 1],
                 color='pink', alpha=0.3, label=" 专 住 95%")

plt.xlabel("转专")
plt.ylabel("专 (专)")
plt.title(f" 转 {ticker} - 21 砖注转 拽 注 驻转 专")
plt.legend()
plt.grid()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

#  专转 转 砖 转 驻 (AAPL) 注 转 砖注
ticker = "AAPL"
#df = yf.download(ticker, start="2020-01-01", end="2025-04-02", interval="1wk")
df = yf.download(ticker, start="2020-01-01", end="2025-04-02")
df.to_csv(f"{ticker}.csv")
df = df[['Close']]  # 砖专 专拽 转 专 转
df.rename(columns={'Close': 'price'}, inplace=True)

#  转转  SARIMA 注 驻专专 (1,0,2) -(0,2,2,52)
order = (2, 1, 0)  # ARIMA (p,d,q)
seasonal_order = (1, 2, 2, 52)  # 注转转 (P,D,Q,m)

model = SARIMAX(df['price'], order=order, seasonal_order=seasonal_order)
result = model.fit()

#   21 砖注转 拽
forecast_period = 21
forecast = result.get_forecast(steps=forecast_period)
conf_int = forecast.conf_int()

#  爪专转 爪专  注专 
forecast_dates = pd.date_range(start=df.index[-1], periods=forecast_period+1, freq='D')[1:]

#  驻住转 专祝 住专 +  + 驻转 专 专-住
plt.figure(figsize=(12,6))
plt.plot(df.index, df['price'], label="专 住专", color='blue')
plt.plot(forecast_dates, forecast.predicted_mean, label="转转 21 砖注转 拽", color='red')

#  驻住转 驻转 专 (95% Confidence Interval)
plt.fill_between(forecast_dates,
                 conf_int.iloc[:, 0], conf_int.iloc[:, 1],
                 color='pink', alpha=0.3, label=" 专 住 95%")

plt.xlabel("转专")
plt.ylabel("专 (专)")
plt.title(f" 转 {ticker} - 21 砖注转 拽 注 驻转 专")
plt.legend()
plt.grid()
plt.show()
```
